```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=df = pd.read_csv (r'C:\Users\capta\Desktop\dl\House_prediction.csv')
df.head()
```

```python
df_mean=df.groupby(['city']).mean()
df_mean
```

```python
array=df_mean.to_numpy()
city=np.array(['Belo Horizonte','Campinas','Porto Alegre','Rio de Janeiro','São Paulo'])
```

```python
plt.plot(city,array[:,0])
plt.xlabel('City')
plt.ylabel('Area')
plt.show()
```

The variation of areas with different cities is shown.

```python
plt.plot(city,array[:,1],label='Rooms')
plt.plot(city,array[:,2],label='Bathrooms')
plt.plot(city,array[:,3],label='Parking Spaces')
plt.xlabel('City')
plt.ylabel('Number')
plt.legend()
plt.show()
```

The variation of number of rooms,bathrooms and parking spaces with different cities is shown

```python
plt.plot(city,array[:,4],label='House Association Tax')
plt.plot(city,array[:,5],label='Rent amount')
plt.plot(city,array[:,6],label='Property Tax')
plt.plot(city,array[:,8],label='Total')

plt.xlabel('City')
plt.ylabel('Amount(in $)')
plt.legend()
plt.show()
```

The variation of house association tax, rent amount, property tax and total amount with different cities is shown.
